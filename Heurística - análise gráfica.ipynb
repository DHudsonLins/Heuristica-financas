```python
SBUX = dados_series['SBUX']
KOO = dados_series['KO']
PEP = dados_series['PEP']
BRK_B = dados_series['BRK-B']
GME = dados_series['GME']
ASML = dados_series['ASML']
LMT = dados_series['LMT']
MCD = dados_series['MCD']
VOO = dados_series['VOO']
```

```python
SBUX.name = 'SBUX'
KOO.name = 'KOO'
PEP.name = 'PEP'
BRK_B.name = 'BRK_B'
GME.name = 'GME'
ASML.name = 'ASML'
LMT.name = 'LMT'
MCD.name = 'MCD'
VOO.name = 'VOO'
```

```python
lista_dataframe_acoes = [SBUX, KOO, PEP, BRK_B, GME, ASML, LMT, MCD, VOO]
```

```python
## VISUALIZAÇÕES PARA DEFINIR MELHORES AÇÕES
```

```python
for df in lista_dataframe_acoes:
    plt.figure(figsize=(12,8))
    sns.set_theme(style="whitegrid")
    sns.distplot(df['Close'].dropna(), color = 'green').set_title(df.name)
    sns.jointplot(x=df['Open'],y=df['Close'], color = 'green',data=df)
    sns.pairplot(df, corner = True)
```

```python
# Observando algumas ações, vemos que temos uma tendência de algumas a permanecerem com um valor fixo durante bastante tempo
# Como a a VOO, a GME, a BRK_B e a ASML
# Enquanto outras tendem a serem muito instáveis como a lMT, a PEP, a KOO, a SBUX e a MCD 
# A LMT tem tendência de quedas bruscas.
# baseado nas visualizações, as 5 melhores carteiras de investimeto seriam: a MCD, pois oscila sempre numa faixa aceitável, a BRK_B pela consistência, a PEP por altos valores quase sempre, a KO e a SBUX pelos mesmos motivos
```
